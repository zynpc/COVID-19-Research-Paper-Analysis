## Analysis of COVID-19 Related Research Papers

In this notebook, we analyze a corpus of COVID-19 related research papers to uncover trends in publication behavior, research activity, and technical content throughout the pandemic. 

### Objectives:
- Understand the temporal growth of COVID-19 research over the years.
- Identify the most active journals and authors contributing to the domain.
- Distinguish between types of publications (preprints vs peer-reviewed journals).
- Extract the most prominent technical terms from abstracts using TF-IDF.
- Assess the overall sentiment of research abstracts to understand the tone of scientific communication during the pandemic.

### Methods Used:
- **Data Cleaning**: Removed duplicates, missing values, and standardized formats.
- **Trend Analysis**: Analyzed the yearly count of publications to observe how COVID-19 research output evolved over time.
- **Exploratory Statistics**: Identified the top 10 most common journals by publication count and highlighted the most prolific authors contributing to the research.
- **Publication Type Classification**: Categorized papers into Preprints and Journal articles based on the `source_x` field to analyze distribution and publication patterns.
- **TF-IDF Analysis**: Extracted significant domain-specific terms.
- **Sentiment Analysis**: Applied polarity scoring to abstracts using TextBlob.

### Summary:
This multi-angle analysis provides insights into how the scientific community responded to COVID-19, both in terms of quantity and content. We observe a sharp increase in research activity post-2020, with key journals and authors emerging as major contributors. TF-IDF revealed dominant terminology related to virology and medicine, while sentiment analysis suggested a generally neutral and positive tone, with variations based on topic and year.

This notebook combines classical statistical techniques with NLP to offer a comprehensive picture of COVID-19 scholarly communication.

## Data Cleaning

```python
# Select only the useful columns
columns_to_keep = ['title', 'abstract', 'publish_time', 'journal', 'authors', 'doi', 'url', 'source_x']
df = df[columns_to_keep]

# Drop rows with missing title or publish date
df = df.dropna(subset=['title', 'publish_time'])

# Convert publish_time to datetime format
df['publish_time'] = pd.to_datetime(df['publish_time'], errors='coerce')

# Drop rows where publish_time conversion failed
df = df.dropna(subset=['publish_time'])

# Drop duplicate titles
df = df.drop_duplicates(subset='title')

df['abstract'] = df['abstract'].astype(str)

# Then, remove rows where abstract is missing or too short
df = df[df['abstract'].str.len() > 20]
```

```python
# Save cleaned data
df.to_csv("metadata_clean.csv", index=False)

# Print final shape
print(f"Cleaned shape: {df.shape}")
```

### Increase in COVID-19 research over the years

```python
import matplotlib.pyplot as plt

# Extract year from publish_time
df['year'] = df['publish_time'].dt.year

# Count publications per year
yearly_counts = df['year'].value_counts().sort_index()


plt.figure(figsize=(10, 6))
plt.plot(yearly_counts.index, yearly_counts.values, marker='o')
plt.title("Number of Publications per Year")
plt.xlabel("Year")
plt.ylabel("Number of Publications")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Most Common Journals

```python
# Count top 10 journals
top_journals = df['journal'].value_counts().head(10)


plt.figure(figsize=(10, 6))
top_journals.plot(kind='bar')
plt.title("Top 10 Journals by Number of Publications")
plt.xlabel("Journal")
plt.ylabel("Number of Publications")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

###  Most Active Authors

```python
from collections import Counter

# Drop missing author data
df_authors = df[df['authors'].notnull()]

# Split authors and flatten list
all_authors = df_authors['authors'].str.split(';|,').explode().str.strip()
top_authors = Counter(all_authors).most_common(10)

# Convert to DataFrame for plotting
authors_df = pd.DataFrame(top_authors, columns=['Author', 'Publication Count'])


plt.figure(figsize=(10, 6))
plt.bar(authors_df['Author'], authors_df['Publication Count'], color='salmon')
plt.title("Top 10 Most Active Authors")
plt.xlabel("Author")
plt.ylabel("Number of Publications")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Publication Types: Preprint vs Journal

```python

def classify_source(source):
    if pd.isnull(source):
        return 'Unknown'
    
    source = source.lower()
    if 'biorxiv' in source or 'medrxiv' in source or 'arxiv' in source:
        return 'Preprint'
    elif 'pmc' in source or 'pubmed' in source or 'springer' in source or 'elsevier' in source or 'wiley' in source:
        return 'Journal'
    else:
        return 'Other'


df['publication_type'] = df['source_x'].apply(classify_source)

# Count publication types
type_counts = df['publication_type'].value_counts()


plt.figure(figsize=(8, 5))
type_counts.plot(kind='bar', color='mediumseagreen')
plt.title("Publication Types: Preprint vs Journal")
plt.xlabel("Type")
plt.ylabel("Number of Publications")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()


  
```
